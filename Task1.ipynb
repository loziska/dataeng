```python
def add_employee(data, emp):
    data.append(emp)

def delete_employee(data, emp_id):
    data.remove(emp_id)

def update_employee(data, emp_id, new_data):
    for emp in data:
        if emp["id"] == emp_id:
            emp.update(new_data)

def print_all(data):
    for emp in data:
        print(emp)
```
